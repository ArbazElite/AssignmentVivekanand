```python
pos_review_files=glob.glob(f'{pos_review_path}/*.txt')
neg_review_files=glob.glob(f'{neg_review_path}/*.txt')
```

```python
review_list=[]
for file_name in pos_review_files:
#     print(file_name)
    f=open(file_name,'r')
    text=f.read()
    text=re.sub('[^A-Za-z]+',' ',text)
    review_list.append(text)
    
for file_name in neg_review_files:
#     print(file_name)
    f=open(file_name,'r')
    text=f.read()
    text=re.sub('[^A-Za-z]+',' ',text)
    review_list.append(text)
    
len(review_list)
```

```python
pos_target=np.ones(len(pos_review_files),dtype=int)
neg_traget=np.zeros(len(neg_review_files),dtype=int)
# neg_traget
y=np.append(pos_target,neg_traget)
y=pd.Series(y)
y
```

```python
tfidf_vector=TfidfVectorizer(lowercase=True,stop_words='english',min_df=10)
x_tfidf_vector=tfidf_vector.fit_transform(review_list)
tfidf_columns=tfidf_vector.get_feature_names()
x_tfidf_df=pd.DataFrame(x_tfidf_vector.toarray(),columns=tfidf_columns)
x_tfidf_df
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_tfidf_df,y,test_size=0.2,random_state=1,stratify=y)
```

```python
# Testing accuracy
mnb_model=MultinomialNB()
mnb_model.fit(x_train,y_train)
y_pred_test=mnb_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred_test)
print('Testing Accuracy :',accuracy)
clf_report=classification_report(y_test,y_pred_test)
print('\n Classificatio Report:\n',clf_report)
cnf_matrix=confusion_matrix(y_test,y_pred_test)
cnf_matrix
```

```python
# Training accuracy
mnb_model=MultinomialNB()
mnb_model.fit(x_train,y_train)
y_pred_train=mnb_model.predict(x_train)
accuracy=accuracy_score(y_train,y_pred_train)
print('Training Accuracy :',accuracy)
clf_report=classification_report(y_train,y_pred_train)
print('\n Classificatio Report:\n',clf_report)
cnf_matrix=confusion_matrix(y_train,y_pred_train)
cnf_matrix
```
